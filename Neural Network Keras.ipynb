```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
import tensorflow as tf
import tensorflow.keras as keras
import warnings
warnings.filterwarnings("ignore")
```

```python
X, Y = make_circles(n_samples=500, factor=0.5, noise=0.05)

res = 100

_x0 = np.linspace(-1.5, 1.5, res)
_x1 = np.linspace(-1.5, 1.5, res)

_pX = np.array(np.meshgrid(_x0, _x1)).T.reshape(-1, 2)
_pY = np.zeros((res,res)) + 0.5

plt.figure(figsize=(8, 8))
plt.pcolormesh(_x0, _x1, _pY, cmap='hot', vmin=0, vmax=1)

plt.scatter(X[Y == 0, 0], X[Y == 0, 1], c='skyblue')
plt.scatter(X[Y == 1, 0], X[Y == 1, 1], c = 'white')
```

```python
lr = 0.01
nn = [2, 16, 8, 1]

model = keras.Sequential()

l1 = model.add(keras.layers.Dense(nn[1], activation='relu'))
l2 = model.add(keras.layers.Dense(nn[2], activation='relu'))
l3 = model.add(keras.layers.Dense(nn[3], activation='sigmoid'))

model.compile(loss='mse', optimizer=keras.optimizers.SGD(lr=0.05), metrics=['acc'])

model.fit(X, Y, epochs=100)
```
